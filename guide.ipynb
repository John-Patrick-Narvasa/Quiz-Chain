```python
from langchain_google_genai import ChatGoogleGenerativeAI as CGA
from dotenv import load_dotenv

load_dotenv()

def generate_output():
    llm = CGA(model="gemini-2.5-flash", temperature=1.0, max_tokens=1000)

    name = llm("I have a dog pet and I want a cool name for it. Suggest me FIVE cool names for my pet.")

    return name.content
if __name__ == "__main__":
    print(generate_output())
```
